```python
#8.1.1 task 1
def sum_nested_list(nested_list):
    total_sum = 0
    for element in nested_list:
        if isinstance(element, list):
            total_sum += sum_nested_list(element)
        else:
            total_sum += element
    return total_sum
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print(result)
```

```python
#task2

def generate_permutations(s):
    result = []
    def backtrack(current_permutation, remaining_chars):
        if len(current_permutation) == len(s):
            result.append("".join(current_permutation))
            return
        for i in range(len(remaining_chars)):
            char = remaining_chars[i]

            if i > 0 and remaining_chars[i] == remaining_chars[i-1]:
                continue
            backtrack(current_permutation + [char], remaining_chars[:i] + remaining_chars[i+1:])


    sorted_s = sorted(s)
    backtrack([], sorted_s)
    return result

print(generate_permutations("abc"))

print(generate_permutations("aab"))
```

```python
#task 3

def calculate_directory_size(directory):
    total_size = 0
    for name, size_or_subdir in directory.items():
        if isinstance(size_or_subdir, int):
            # It's a file, add its size
            total_size += size_or_subdir
        elif isinstance(size_or_subdir, dict):
            # It's a subdirectory, recurse
            total_size += calculate_directory_size(size_or_subdir)
    return total_size
sample_directory = {
    "file1.txt": 100,
    "file2.txt": 50,
    "subdir1": {
        "file3.txt": 200,
        "file4.txt": 150,
        "subdir2": {
            "file5.txt": 75
        }
    },
    "file6.txt": 300
}
total_size_kb = calculate_directory_size(sample_directory)
print(f"The total size of the directory is: {total_size_kb} KB")
```

```python
# 8.2.1 Task 1

def min_coins(coins, amount):

    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for i in range(1, amount + 1):

        for coin in coins:

            if i >= coin:

                dp[i] = min(dp[i], dp[i - coin] + 1)
    if dp[amount] == float('inf'):
        return -1
    else:
        return dp[amount]
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
```

```python
# task 2
def longest_common_subsequence(s1, s2):
    # Get the lengths of the strings
    m = len(s1)
    n = len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m]
s1 = "abcde"
s2 = "ace"
lcs_length = longest_common_subsequence(s1, s2)
print(f"The length of the Longest Common Subsequence for '{s1}' and '{s2}' is: {lcs_length}")
```

```python
# Task 3
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    return dp[capacity]
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight = weights[i - 1]
        value = values[i - 1]
        for w in range(1, capacity + 1):
            if weight <= w:
                dp[i][w] = max(value + dp[i - 1][w - weight], dp[i - 1][w])
            else:

                dp[i][w] = dp[i - 1][w]
    return dp[capacity]

weights_input = [1, 3, 4, 5]
values_input = [1, 4, 5, 7]
capacity_input = 7
result = knapsack(weights_input, values_input, capacity_input)
print(f"The maximum value is: {result}")
```
