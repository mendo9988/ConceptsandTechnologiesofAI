```python
time_data = [ (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0), (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5), (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0), (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0), (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0) ]

low_study = []      # less than 3 hours
moderate_study = [] # between 3 and 5 hours (inclusive)
high_study = []     # more than 5 hours

for day in time_data:
    study_hours = day[0]  # study hours is the first element in each tuple
    
    if study_hours < 3:
        low_study.append(study_hours)
    elif 3 <= study_hours <= 5:
        moderate_study.append(study_hours)
    else:  # study_hours > 5
        high_study.append(study_hours)

    print("Low Study Time (less than 3 hours):", low_study)
    print("Moderate Study Time (3-5 hours):", moderate_study)
    print("High Study Time (more than 5 hours):", high_study)
    
```
